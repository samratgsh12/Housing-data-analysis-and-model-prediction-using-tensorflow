```python
X=df.drop('price',axis=1).values
y=df['price'].values
#.values returns the values as numpy arrays rather than pandas data series
#numpy arrays are easily executed by tensorflow
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train) #transforming and fitting done in same step
X_test=scaler.fit_transform(X_test) # no fitting is done so cause we dont want to assume prior information about our test set
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model= Sequential()
#relu- rectified linear unit
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse') #mse-mean squared error(it is used as it is a regression problem)
```

```python
#validation data continously checks the result with the training data
#As the data is larger so we need to pass the data in batches of powers of 2
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
losses =pd.DataFrame(model.history.history)
```

```python
losses.plot() # as the validation loss is going down we can continue training without overfitting
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predictions=model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions) #gives how far of the variance is from the actual value as a rank from 0 to 1
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
```
